```python
import pandas as pd

# Step 1: Data Ingestion 
df = pd.read_csv('spotify_history.csv')
```

```python
df
```

```python
#drop duplicates
df.duplicated()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates()
```

```python
# Step 3: Handle Missing Values
df['reason_start'].fillna('unknown', inplace=True)
df['reason_end'].fillna('unknown', inplace=True)
df
```

```python
df
```

```python
#rename
df.rename(columns={'ts':'Timestamp','ms_played':'Duration_ms','track_name': 'track',
    'artist_name': 'artist',
    'album_name': 'album',
    'reason_start': 'start_reason',
    'reason_end': 'end_reason'},inplace=True)
```

```python
df
```

```python
#data type correction
df['Timestamp']=pd.to_datetime(df['Timestamp'])
```

```python
df
```

```python
df[df['Timestamp']>'2023-11-01']
```

```python
df['2024'] = df['Timestamp'].dt.year
#extracting year
df
```

```python
# Format Standardization
text_columns = ['platform', 'track', 'artist', 'album', 'start_reason', 'end_reason']
for col in text_columns:
    df[col] = df[col].astype(str).str.strip().str.upper()
```

```python
df
```

```python
df.to_csv('cleaned_spotify_history.csv', index=False)
```
